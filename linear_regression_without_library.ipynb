```python
import numpy as np

# Create sample data
np.random.seed(42)
X = np.arange(1, 11)          # x = [1,2,3,...,10]
Y = 3 * X + 4 + np.random.randn(10) * 2  # add small noise
```

```python
def linear_regression(X, Y):
    # Convert to 2D column vector
    X = X.reshape(-1, 1)
    
    # Add bias column (ones)
    X_b = np.c_[np.ones((X.shape[0], 1)), X]  # shape: (n_samples, 2)
    
    # Apply Normal Equation
    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)
    
    intercept = theta[0]
    slope = theta[1]
    
    return intercept, slope
```

```python
intercept, slope = linear_regression(X, Y)
print(f"Learned Equation: y = {slope:.2f}x + {intercept:.2f}")
```

```python
def predict(X, slope, intercept):
    return slope * X + intercept

Y_pred = predict(X, slope, intercept)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Predicted Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
